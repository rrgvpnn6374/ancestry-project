## 1. Introduction & Problem Statement
**The Problem:** Genetic data is high-dimensional and unlabeled in many biological contexts. A key challenge in population genetics is identifying substructures within a population without prior knowledge of ancestry.
**The Goal:** The goal of this project is to build an unsupervised learning pipeline that can:
1.  Clean and transform raw genotype data (nucleotides).
2.  Reduce dimensionality to visualize population structure.
3.  Cluster individuals into super-populations using **K-Means** and **Hierarchical Clustering**.

## 2. Data Collection and Provenance
* **Source:** The data is a subset of the **1000 Genomes Project (Phase 3)**, an international research effort to establish the most detailed catalogue of human genetic variation.
* **Dataset:** I utilized an educational subset of the data containing approximately 2,504 individuals and ~10,000 Single Nucleotide Polymorphisms (SNPs).
* **Provenance:** The data was collected via whole-genome sequencing of anonymous volunteers who provided informed consent. The subset used here was pre-processed to include Ancestry Informative Markers (AIMs).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import io
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage

# --- LOAD DATA ---
url = "https://raw.githubusercontent.com/sakshamg94/PCA-genome-data/master/p4dataset2020.txt"
print("Downloading data... (This may take a moment)")
s = requests.get(url).content

# Read space-separated text file
df = pd.read_csv(io.StringIO(s.decode('utf-8')), sep="\s+", header=None)

# Separate Metadata (Cols 0-2) from Genotypes (Cols 3+)
metadata = df.iloc[:, :3]
metadata.columns = ['sample_id', 'gender', 'population']

# Isolate the DNA data
genotypes_raw = df.iloc[:, 3:]

print(f"Data Loaded! Samples: {df.shape[0]}, SNPs: {genotypes_raw.shape[1]}")
print(f"Example raw data: {genotypes_raw.iloc[0, 0]}") # Should show a letter like 'G' or 'A'
```

## 3. Exploratory Data Analysis (EDA) & Cleaning
Genetic data consists of nucleotides (A, T, G, C) which are categorical. To perform mathematical operations (like PCA), these must be converted into numerical values.

**Data Transformation Strategy:**
* **Encoding:** I converted the categorical data into a binary format (0 and 1) based on the **Major Allele**. If an individual has the most common variant, they are assigned `0`; if they have the variant (minor allele), they are assigned `1`.
* **Minor Allele Frequency (MAF) Filtering:** Rare variants (mutations found in only a few people) act as noise in clustering algorithms. I analyzed the distribution of allele frequencies and filtered out SNPs with a frequency below 5% to focus on common variants that drive population differences.

```python
# --- STEP 1: CONVERT LETTERS TO NUMBERS ---
print("Transforming data from Letters to Numbers...")

# Find the most common letter (Mode) for each column
modes = genotypes_raw.mode().iloc[0]

# Convert: Match = 0 (Normal), Mismatch = 1 (Variant)
genotypes_numeric = (~(genotypes_raw.eq(modes))).astype(int)

print("Transformation Complete.")

# --- STEP 2: MAF FILTERING (EDA) ---
# Calculate frequency of the variant (the '1's)
n_people = genotypes_numeric.shape[0]
allele_freqs = genotypes_numeric.sum(axis=0) / n_people 

# PLOT: Histogram of Allele Frequencies
plt.figure(figsize=(10, 6))
plt.hist(allele_freqs, bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Allele Frequencies (Before Filtering)')
plt.xlabel('Frequency of Variant')
plt.ylabel('Count of SNPs')
plt.show()

# FILTER: Keep only SNPs with frequency > 5%
maf_threshold = 0.05
snps_to_keep = allele_freqs[(allele_freqs > maf_threshold) & (allele_freqs < (1 - maf_threshold))].index
genotypes_clean = genotypes_numeric[snps_to_keep]

print(f"Original SNP count: {genotypes_numeric.shape[1]}")
print(f"Final Clean SNP count: {genotypes_clean.shape[1]}")
```

## 4. Feature Engineering: Dimensionality Reduction
The "Curse of Dimensionality" is a significant hurdle in genetics, as the number of features (SNPs) far exceeds the number of samples.

**Method: Principal Component Analysis (PCA)**
I applied PCA to reduce the feature space from ~4,000 dimensions to 2 principal components.
* **Scaling:** Data was standardized using `StandardScaler` to ensure all SNPs contributed equally to the variance.
* **Interpretation:** The scatter plot below visualizes the 2,504 individuals projected onto the first two principal components. The distinct "clouds" suggest that the genetic variance is strongly correlated with ancestral background.

```python
# --- RUN PCA ---
# 1. Standardize the data (Mean=0, Variance=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(genotypes_clean)

# 2. Compute PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 3. Plot PCA results with Population Labels
plt.figure(figsize=(12, 8))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=metadata['population'], palette='tab10', s=50, alpha=0.8)
plt.title('PCA of Genetic Data (True Labels)')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} Variance)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} Variance)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## 5. Unsupervised Model Building: K-Means Clustering
To formally identify the groups visible in the PCA plot, I selected **K-Means Clustering**.

**Hyperparameter Optimization (The Elbow Method):**
Determining the optimal number of clusters ($K$) is critical. I used the Elbow Method, plotting the Inertia (sum of squared distances to the nearest cluster center) against $K$.
* **Observation:** The plot shows a distinct "elbow" at **K=3**, indicating that adding more clusters beyond 3 yields diminishing returns in explaining the variance.

**Model Evaluation:**
I trained the K-Means model with $K=3$ and evaluated the performance using the **Silhouette Score**. A score of **0.573** indicates that the clusters are well-separated and dense.

```python
# --- ELBOW METHOD ---
inertia = []
range_k = range(1, 10)

for k in range_k:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_pca)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(range_k, inertia, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# --- FINAL K-MEANS MODEL (K=3) ---
k_optimal = 3
kmeans_final = KMeans(n_clusters=k_optimal, random_state=42)
cluster_labels = kmeans_final.fit_predict(X_pca)

# Visualize the Clusters (What the computer found)
plt.figure(figsize=(12, 8))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=cluster_labels, palette='viridis', s=50)
centers = kmeans_final.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title(f'K-Means Clustering Results (K={k_optimal})')
plt.legend(title='Cluster ID')
plt.show()

# Evaluation Score
print(f"Silhouette Score: {silhouette_score(X_pca, cluster_labels):.3f}")
```

## 6. Model Comparison: Hierarchical Clustering
While K-Means forces data into distinct "blobs," human evolution is a continuous branching process. To capture this, I compared K-Means with **Hierarchical Clustering**.

**Analysis of Dendrogram:**
The Dendrogram (tree diagram) below reveals the hierarchical structure of the data:
1.  **First Split:** The tree splits into two major branches at a high distance, likely separating African populations from Non-African populations (the oldest divergence).
2.  **Second Split:** The right branch splits again, separating European/Asian populations.
3.  **Validation:** Cutting the tree at a distance of 150 (dashed line) results in exactly **3 clusters**, validating the choice of $K=3$ derived from the K-Means Elbow method.


```python
# --- HIERARCHICAL CLUSTERING ---
# Compute the Linkage (The Family Tree)
Z = linkage(X_pca, method='ward')

# Plot the Dendrogram
plt.figure(figsize=(12, 6))
dendrogram(Z, truncate_mode='lastp', p=30, leaf_rotation=90., leaf_font_size=10., show_contracted=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
plt.axhline(y=150, c='k', ls='--', lw=1) # Visual cut-off line
plt.show()
```
